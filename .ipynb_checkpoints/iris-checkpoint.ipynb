```python
from sklearn.datasets import load_iris
iris_dataset = load_iris() #It contains certain keys and values in a Bunch object (similare to a dictionary)
```

```python
print(f"Keys of iris dataset: \n{iris_dataset.keys()}")
```

```python
# The value of the key DESCR is a short description of the dataset
print(iris_dataset['DESCR'][:193] + "\n...")
```

```python
# The value of the key "target_names" is an array of strings containing the species of the flower we want to predict
print(f"Target names: {iris_dataset['target_names']}")
```

```python
# The value of the key "feature_names" is a list of strings, giving the description of each feature
print(f"Feature names: \n{iris_dataset['feature_names']}")
```

```python
print(f"Type of data: {type(iris_dataset['data'])}")
```

```python
print(f"Shape of data: {iris_dataset['data'].shape}")
```

```python
# Here we see that the array contains the measurements of 150 different flowers
# Individual items --> samples
# Properties of the samples --> features
# Here (150,4) means there are 150 samples, each with 4 features {intution}
```

```python
# Each column in the dataset represents the features 
# So, there would be 4 columns for the sepal length, sepal width, petal length, petal width

print(f"Data for the first five flowers: \n{iris_dataset['data'][0:5]}")
```

```python
# From the above data, we can say that:
# first five flower has the same petal width of 0.2 cm
# first flower has the longest sepal which is 5.1 cm
```

```python
# "target" array contains the species of each flower that were measured

print(f"Type of target: {type(iris_dataset['target'])}")
```

```python
# Here, target is one dimensional array with one entry per flower
print(f"Shape of target: {iris_dataset['target'].shape}")
```

```python
# The species are encoded as integers from 0 to 2
print(f"Target: \n{iris_dataset['target']}")

# 0--> setosa
# 1--> versicolor
# 2 --> virginica
```

```python
# We need to split the data into two parts: training set and testing set
# y =f(x), for cetain values of X, we get values of y
# Here, X means input set and y means output set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)

# This splits the whole data:
# 75% to the training set --> X_train, y_train
# 25% to the test set --> X_test, y_test
# random_state=0 shuffles the whole data while doing the spliting
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
# create dataframe from data in X_train
# label the columns using the strings in iris_dataset.feature_names

import pandas as pd
import mglearn

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# create a scatter matrix from the dataframe, color by y_train

grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10,10), marker='o', hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
# k-nearest neighbours classification algorithm Pg-20

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)

# The knn object encapsulates the algorithm that will be used to build the model from the training data as well the algorithm to make predictions on new data points
# Holds the info that the algorithm has extracted from the training set
```

```python
# To build the model on the training set

knn.fit(X_train, y_train)

# fit method returns the knn object itself
```

```python
# Making Predictions
# Suppose we get an iris with:
# sepal length = 5cm
# sepal width = 2.9 cm
# petal length = 1 cm
# petal width = 0.2 cm

import numpy as np

X_new = np.array([[5,2.9,1,0.2]]) # 2-dimensional array
print("X_new.shape: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))

# Here we see, our model predicts that this new iris belongs to the class 0, meaning its species is sertosa
```

```python
# Evaluating the model
# test set will be used to evaluate the model

y_pred = knn.predict(X_test)
print("Test set predictions: \n{}".format(y_pred))
```

```python
print("Test Set Score: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
# Thus, for this model, the test set accuracy is about 0.97
# It means the model makes 97% correct prediction about the test set w.r.t the train set
```
